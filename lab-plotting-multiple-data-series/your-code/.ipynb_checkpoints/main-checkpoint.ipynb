# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from matplotlib import dates

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head(4)
```

```python
data.shape
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
sales = data.groupby('ItemType')[['RetailSales', 'RetailTransfers', 'WarehouseSales']].sum()
sales.head()
```

```python
sns.set(style='darkgrid')
sales.plot.bar(figsize=(15, 5))
plt.title('Sales by Item Type', size=22)
plt.show();
```

```python
sales.plot.bar(figsize=(15, 5))
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data['total_sales'] = data['RetailSales'] + data['WarehouseSales']
data.head()
```

```python
data_total_sales = data.groupby('Supplier')[['RetailSales', 'WarehouseSales', 'total_sales']].sum().sort_values(by=['total_sales'], ascending=False)
data_total_sales.head()
```

```python
# Opcin 1
data_total_sales.head(10).sort_values(by=['total_sales']).plot.barh()
plt.title('Top 10 suppliers by sales', size=15);
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
sales = data.groupby('Month')[['RetailSales', 'RetailTransfers', 'WarehouseSales']].mean()
sns.lineplot(data=sales, dashes=False)
plt.title('Average Sales per month', size=15)
plt.xticks(range(1,13))
plt.show();
```

## 4. Plot the same information as above but as a bar chart.

```python
# Opcin 1
sales.plot.bar()
plt.title('Average Sales per month', size=15)
plt.show();
```

```python
# Opcin 2
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

sns.barplot(x='Month', y='RetailSales', data=data, palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Retail Sales")

sns.barplot(x='Month', y='RetailTransfers', data=data, palette="vlag", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Retail Transfers")

sns.barplot(x='Month', y='WarehouseSales', data=data, palette="deep", ax=ax3)
ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("Warehouse Sales");
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
# Combine year and month columns

# Agregamos una columna con el da (usando .assign(), para pasarlo todo a una fecha
data['Date'] = pd.to_datetime(data[['Year', 'Month']].assign(Day=1))
data.head(4)
```

```python
data_chart = data.pivot_table(values = 'RetailSales', columns = 'ItemType', index ='Date', aggfunc = sum).fillna(0)
data_chart.head(4)
```

```python
data_chart.plot(figsize=(13, 8))
plt.legend(loc='upper right')
plt.title('Sales by Item Type over Time', size=17)
plt.show();
```

## 6. Plot the same information as above but as a bar chart.

```python
data_chart.plot.bar(figsize=(13, 8));
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
sns.lmplot(data=data, x='RetailSales', y='RetailTransfers', hue='ItemType')
plt.title('Sales vs Transfers by Item Type', size=15)
plt.xlim(left=-100)
plt.ylim(bottom=-100)
plt.show();
```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.

*Hint: Seaborn's pairplot may be your best option here.*

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
num_data = data.select_dtypes(include=numerics)
num_data.head(4)
```

```python
num_data['ItemType'] = data['ItemType']
num_data.head(4)
```

```python
sns.pairplot(data=num_data, hue='ItemType');
```
